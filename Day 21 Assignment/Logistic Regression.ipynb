```python
dataset.dropna(inplace=True)
dataset.shape
```

```python
dataset.columns
```

```python
dataset = dataset.drop(columns = ['EmployeeID', 'EmployeeCount', 'Over18', 'StandardHours', ])
```

```python
dataset.info()
```

```python
from sklearn import preprocessing
a = preprocessing.LabelEncoder()
dataset['Attrition'] = a.fit_transform(dataset['Attrition'])
dataset['BusinessTravel'] = a.fit_transform(dataset['BusinessTravel'])
dataset['EducationField'] = a.fit_transform(dataset['EducationField'])
dataset['Gender'] = a.fit_transform(dataset['Gender'])
dataset['Department'] = a.fit_transform(dataset['Department'])
dataset['MaritalStatus'] = a.fit_transform(dataset['MaritalStatus'])
dataset['JobRole'] = a.fit_transform(dataset['JobRole'])
```

```python
Y1 = dataset['Attrition']
X2 = dataset.drop(columns = ['Attrition'])
```

```python
X3 = sm.add_constant(X2)
Logistic1 = sm.Logit(Y1,X3)
```

```python
res = Logistic1.fit()
```

```python
res.summary()
```

The following factors highly affect attrition in employees:<br>
1. Age of the employee<br>
2. Marital status of the employee<br>
3. Total number of companies the employee has worked for<br>
4. Total number of years the employee has worked so far<br>
5. Number of times training was conducted for this employee last year<br>

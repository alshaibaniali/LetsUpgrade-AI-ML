# Multiple Linear Regression:-

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('dataset/Linear Regression.xlsx')
data.head()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace = True)
```

```python
data.shape
```

```python
sns.boxenplot(data.price)
```

```python
data.boxplot()
```

```python
sns.pairplot(data)
```

```python
x = data.iloc[:, :1]
y = data.iloc[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
y_train, y_test, x_train, x_test = train_test_split(y ,x, test_size = 0.2, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
mul_reg = LinearRegression()
```

```python
mul_reg.fit(y_train, x_train)
```

```python
mul_reg.coef_
```

```python
mul_reg.intercept_
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
pred = mul_reg.predict(y_test)
```

```python
RMSE = np.sqrt(mean_squared_error(x_test, pred))
r_square = r2_score(x_test, pred)

print('The R^2 value is:', r_square)
print('The RMSE value is:', RMSE)
```
